```python
import numpy as np
import pandas as pd
import scipy 
import matplotlib.pyplot as plt
%matplotlib inline
import statsmodels
from numpy import median
import seaborn as sns
```

 # BAR PLOT

```python
tips = sns.load_dataset('tips')
tips[:5]
```

```python
# black line on bars are error bars
sns.barplot(x='day',y = 'tip', data=tips,hue='sex',palette='spring',order=['Sat','Fri','Thur','Sun','Tue'])
```

```python
sns.barplot(x='day',y = 'total_bill', data=tips,estimator=median, palette='spring')
```

```python
sns.barplot(x='tip',y = 'day', data=tips)
```

```python
sns.barplot(x='smoker',y = 'tip', data=tips,hue = 'sex', estimator=median,palette='spring')
```

```python
sns.barplot(x='smoker',y = 'tip', data=tips)
```

```python
sns.barplot(x='day',y = 'total_bill', data=tips, palette='spring', ci=54)
#ci reducess  the error bar size
```

```python
sns.barplot(x='day',y = 'total_bill', data=tips, palette='spring', ci=56,capsize=.4)
```

```python
sns.barplot(x='day',y = 'total_bill', data=tips,hue='sex', palette='spring', ci=96,capsize=.1)
```

```python
sns.distplot(label_dist,color = 'g',vertical=True)
```

```python
sns.distplot(label_dist,hist=False)
```

# STRIP PLOT

```python
tips = sns.load_dataset('tips')
```

```python
sns.stripplot(x='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=False)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=.1)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, hue='sex', jitter=False,linewidth=1)
```

```python
sns.stripplot(x='day',y='total_bill',hue='sex', linewidth=1, data=tips, jitter=False, split=True)
```

```python
sns.stripplot(x='day', y='total_bill',hue='sex', edgecolor='gray',data=tips, marker='*',size=10,jitter=False)
```

```python
sns.stripplot(x='day', y='total_bill', edgecolor='gray',alpha=0.5,data=tips, marker='*',size=10,jitter=False,split=True,hue='sex')
```

```python
sns.stripplot(x='tip',y='day',data=tips,jitter=True,palette='winter_r',color=.1)
sns.boxplot(x='tip',y='day',data=tips)
```

```python
sns.stripplot(x='tip',y='day',data=tips,jitter=True,palette='winter_r',color=.1)
sns.violinplot(x='tip',y='day',data=tips, color='0.9')
```
